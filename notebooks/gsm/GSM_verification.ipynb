```python
import numpy as np
n = 10000
p=10
s = np.random.poisson(size=n, lam=p)
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
y = []
j = 11
for i in range(n-j):
    y.append(np.sum(s[i:i+j]))
y=np.array(y)
```
